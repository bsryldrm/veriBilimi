##Prosedür 2 – Özellik Mühendisliği

Daha iyi analiz için bazı özelliklerin değiştirilmesi veya optimize edilmesi gerekir. Bakalım yolcu aileleri ne kadar büyük? Aile büyüklüğünü görmek için SibSp ve Parch özelliklerini birleştirebiliriz, çünkü SibS gemideki kardeşlerin/eşlerin sayısını tutar ve Parch gemideki ebeveynlerin/çocukların sayısıyla ilgilidir.

```python
import pandas as pd

titanic = pd.read_csv('/content/drive/MyDrive/YapayZekaUzmanı/veribilimiders/ders_2/titanic.csv')
```

```python
titanic["FamilySize"] = titanic["SibSp"] + titanic["Parch"]
print(titanic["FamilySize"].value_counts())
```

Şimdi aile büyüklüğünü dört gruba ayırın; bekar, sadece eş, küçük aile (5'ten az) ve geniş aile (5 veya daha fazla aile üyesi).

```python
titanic.loc[titanic["FamilySize"] == 0, "FsizeD"] = 'singleton'
titanic.loc[titanic["FamilySize"] == 1, "FsizeD"] = 'spouse'
titanic.loc[(titanic["FamilySize"] > 1)  &
            (titanic["FamilySize"] < 5), "FsizeD"]  = 'small'
titanic.loc[titanic["FamilySize"] > 4, "FsizeD"] = 'large'

print(titanic["FsizeD"].unique())
print(titanic["FsizeD"].value_counts())
```
