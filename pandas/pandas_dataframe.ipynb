# Pandas DataFrame ile Groupby Kullanımı

# Pandas ile kayıp ve bozuk veri analizi

```python
import pandas as pd
import numpy as np
```

```python
data = np.random.randint(10,100,15).reshape(5,3)
```

```python
df = pd.DataFrame(data, index = ['a','c','e', 'f','h'], columns =['Column1', 'Column2', 'Column3'] )
df = df.reindex(['a','b','c','d','e','f','g','h'])
df
```

```python
df.drop("Column1", axis = 1)
```

```python
df.drop('a',axis = 0)
```

```python
df.isnull()  # boş mu
```

```python
df.isnull().sum()
```

```python
df["Column1"].isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(how = "any")
```

```python
df.dropna(axis=1)  # sütuna göre
```

```python
df.dropna(thresh= 1)
```

```python
df.fillna(value = 'no input')
```

```python
df.fillna(value = 1)
```

```python
def ortalama(df):
  toplam = df.sum().sum()
  adet = df.size - df.isnull().sum().sum()

  return toplam /adet    # 59.2

df.fillna(value = ortalama(df))
```

# Pandas ile Join ve Merge İşlemleri

```python
import pandas as pd
```

```python
customers = {
      'CustomerId' : [1,2,3,4],
      "firstname" : ["Ahmet", "Ali", "Hasan", "Canan"],
      "lastname" : ["Yılmaz", "Korkmaz", "Çelik", "Toprak"]
}
```

```python
orders = {
    "OrderId" : [10,11,12,13],
    "CustomerId" : [1,2,5,7],
    "OrderDate" : ["2010-07-04", "2010-08-04", "2010-07-07", "2012-07-04"]
}
```

```python
df_customers = pd.DataFrame(customers, columns = ["CustomerId", "firstname", "lastname"])
df_orders = pd.DataFrame(orders, columns = ["CustomerId", "OrderId", "OrderDate"])
```

```python
df_customers
```

```python
df_orders
```

```python
pd.merge(df_customers,df_orders,how="inner")
```

```python
pd.merge(df_customers,df_orders,how="left")
```

```python
pd.merge(df_customers,df_orders,how="right")
```

```python
customersA = {
      'CustomerId' : [1,2,3,4],
      "firstname" : ["Ahmet", "Ali", "Hasan", "Canan"],
      "lastname" : ["Yılmaz", "Korkmaz", "Çelik", "Toprak"]
}
```
